Comprehending the information from the given data


    let us understand in the way hierarchical order of features in the data, route_schedule_uuid is the unique id for each route travelled by load carriers. Trip_uuid is the unique id for every journey travelled by load carriers. In each route_scheduled_uuid there are several trips done in the given period. For every trip we have trip_creation_time, which is the datetime stamp of trip created, od_start_time is the trip start datetime stamp, od_end_time is the trip end datetime stamp, start_scan_to_end_scan is the time taken for the trip which is difference of start time and end time. Actual time and actual_distance_to_destination are the recorded values of the trip checkpoints in between the journey and values are in cumulatives. osrm_time, osrm_distance are the values given by machine, are also in cumulative form. Segment_actual_time, segment_osrm_time, segment_osrm_distance are the time duration taken between the checkpoints of a parcel.

Structuring the data in such a way each record represents for unique trip_uuid by 2 steps

Step1:

    1.Evaluating the numerical features for each unique source to destination in each trip_uuid by grouping the columns of data, trip_creation_time, route_schedule_uuid, route_type, trip_uuid, source_center, source_name, destination_center, destination_name, od_start_time, od_end_time, start_scan_to_end_scan and aggregating the columns actual_distance_to_destination, actual_time, osrm_time, osrm_distance, segment_actual_time, segment_osrm_time, segment_osrm_distance with max, max, max, max, sum, sum, sum respectively. For each trip_uuid sorting its details by od_start_time this is like ordering source name and destination name in such a way that source name will be its previous row destination name of its trip_uuid. 

```python
grouping_columns=['data', 'trip_creation_time', 'route_schedule_uuid', 'route_type',
       'trip_uuid', 'source_center', 'source_name', 'destination_center',
       'destination_name', 'od_start_time', 'od_end_time',
       'start_scan_to_end_scan']
df1=df.groupby(grouping_columns).aggregate({'actual_distance_to_destination':'max','actual_time':'max','osrm_time':'max',
                                        'osrm_distance':'max','segment_actual_time':'sum','segment_osrm_time':'sum',
                                       'segment_osrm_distance':'sum'}).sort_values(by=['trip_uuid','od_start_time'],
                                                                                  ascending=[True,True]).reset_index()
df1
```

Step2:



    Now grouping the columns of data, trip_creation_time, route_schedule_uuid, route_type, trip_uuid and taking first source_center, source_name and od_start_time, last destination_center, destination_name and od_end_time and remainig all will be sum aggregation. 

```python
trips=df1.groupby(grouping_columns[:5]).aggregate({'source_center':'first','source_name':'first',
                                                   'destination_center':'last',
                                            'destination_name':'last','od_start_time':'first','od_end_time':'last',
                                            'start_scan_to_end_scan':'sum','actual_distance_to_destination':'sum',
                                             'actual_time':'sum','osrm_time':'sum','osrm_distance':'sum',
                                             'segment_actual_time':'sum','segment_osrm_time':'sum',
                                             'segment_osrm_distance':'sum'
                                            }).reset_index().sort_values(by='trip_creation_time',ascending=True)
trips
```

About trips dataframe:


        Trips dataframe is the information about every trip's creation_time, source, destination, start time, end_time, trip duration, actual_distance, actual_time, osrm_time, osrm_distance, segment_actual_time, segment_osrm_time, segment_osrm_distance along with their route_schedule_uuid, route_type

Creating New Features 


            From trip_creation_time extracting date
            From source_name extracting city, place_code, State
            From destination_name extracting city, place_code, State
            
            
            Missing values of source name and destination name are filled with their source_center and destination_center so that for these rows same source_center is recorded as state, city, place_code.
            
            In some values of source_name and destination_name, place and code is not mentioned to those place_code is filled with their city names.
            
            In some values of source_name and destination_name, place is not mentioned but code is given to those rows place_code is filled with their codes.

```python
trips['date']=trips['trip_creation_time'].dt.date
trips['date']=pd.to_datetime(trips['date'])
```

```python
def state(x):
    l=x.split('(')
    if len(l)>1:
        return l[-1][:-1]
    else:
        return l[0]
def city(x):
    l=x.split('(')
    l1=l[0].split('_')
    return l1[0]
def place_code(x):
    l=x.split('(')
    l1=l[0].split('_')
    if len(l1)>1:
        return '_'.join(l1[1:])
    else:
        return l1[0]
trips['source_state']=trips['source_name'].apply(state)
trips['source_city']=trips['source_name'].apply(city)
trips['source_place_code']=trips['source_name'].apply(place_code)
trips['destination_state']=trips['destination_name'].apply(state)
trips['destination_city']=trips['destination_name'].apply(city)
trips['destination_place_code']=trips['destination_name'].apply(place_code)
```

Creating IQR function which returns lower limit and upper limit to the given numerical feature.

```python
def IQR(x):
    p25=np.percentile(x,25)
    p75=np.percentile(x,75)
    iqr=p75-p25
    ll=p25-1.5*iqr
    ul=p75+1.5*iqr
    if ll<min(x):
        ll=min(x)
    if ul>max(x):
        ul=max(x)
    return ll,ul
IQR(trips['osrm_distance'])[0]
```

counting the number of values which are outliers in the numerical features of trips dataframe.

```python
for i in ['start_scan_to_end_scan','actual_time','osrm_time','segment_actual_time','segment_osrm_time',
          'actual_distance_to_destination','osrm_distance','segment_osrm_distance']:
    lo=IQR(trips[i])[0]
    up=IQR(trips[i])[1]
    ans=(trips[(trips[i]<lo) | (trips[i]>up)].shape[0])/trips.shape[0]
    print('outliers percentage in',i,round(ans*100,2))
```

All numerical features, outliers percentage are near to 10% so changing the outlier values to their upper limit value.

```python
for i in ['start_scan_to_end_scan','actual_time','osrm_time','segment_actual_time','segment_osrm_time',
          'actual_distance_to_destination','osrm_distance','segment_osrm_distance']:
    up=IQR(trips[i])[1]
    trips.loc[(trips[i]>up), i]=up
```

Normalization of numerical features

```python
def Normalize(x):
    l=min(x)
    u=max(x)
    return (np.array(x)-l)/(u-l)
def Standardize(x):
    mu=np.mean(x)
    sig=np.std(x)
    return (np.array(x)-mu)/sig
```

```python
for i in ['start_scan_to_end_scan','actual_time','osrm_time','segment_actual_time','segment_osrm_time',
          'actual_distance_to_destination','osrm_distance','segment_osrm_distance']:
    st1='normalize'+'_'+i
    st2='standardize'+'_'+i
    trips[st1]=Normalize(trips[i])
    trips[st2]=Standardize(trips[i])
```
